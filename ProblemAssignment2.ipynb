```python
import matplotlib.pyplot as plt

# Data
years = [1810, 1820, 1830, 1840, 1850, 1860, 1870, 1880, 1900, 1910, 1920, 1930, 1940, 1950, 1960, 1970, 1980]
heights = [169.7, 169.1, 166.7, 166.5, 165.6, 166.6, 167.2, 168, 169.4, 170.9, 171, 173.9, 174.9, 176, 176.9, 177.1, 176.8]

# Calculate the mean height
mean_height = sum(heights) / len(heights)

# Output the mean height
print(f"Mean Height: {mean_height:.2f} cm")

# Visualization
plt.figure(figsize=(10, 6))
plt.plot(years, heights, marker='o', color='b')
plt.axhline(y=mean_height, color='r', linestyle='--', label=f'Mean Height: {mean_height:.2f} cm')
plt.title('Trend of Male Heights in the UK (1800s to 1980)')
plt.xlabel('Year')
plt.ylabel('Height (cm)')
plt.grid(True)
plt.legend()
plt.show()
```
